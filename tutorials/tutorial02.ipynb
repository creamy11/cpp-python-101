## `numpy.poly1d` Example
- https://docs.scipy.org/doc/numpy/reference/generated/numpy.poly1d.html

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 5 * x + 1
x = np.linspace(0, 10, 30)
y = f(x)
print(x)
print(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
f = np.poly1d([5, 1])
x = np.linspace(0, 10, 30)
y = f(x)
print(x)
print(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
f = np.poly1d([5, 1])
x = np.linspace(0, 10, 30)
y = f(x) + 4*np.random.normal(size=len(x)) # add noise
xn = np.linspace(0, 10, 200)
plt.plot(x, y, 'or')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
f = np.poly1d([5, 1])
x = np.linspace(0, 10, 30)
y = f(x) + 4*np.random.normal(size=len(x)) # add noise
xn = np.linspace(0, 10, 200)

m, c = np.polyfit(x, y, 1) # Linear regression
yn = np.polyval([m, c], xn)
plt.plot(x, y, 'or')
plt.plot(xn, yn)
plt.grid()
plt.show()
```
