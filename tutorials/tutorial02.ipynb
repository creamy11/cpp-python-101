## `numpy.poly1d` Example
- https://docs.scipy.org/doc/numpy/reference/generated/numpy.poly1d.html

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 5 * x + 1
x = np.linspace(0, 10, 30)
y = f(x)
print(x)
print(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
f = np.poly1d([5, 1])
x = np.linspace(0, 10, 30)
y = f(x)
print(x)
print(y)
```

## Least Square Problem(Linear Regression)
1. Prediction Model : `f(x)`
    - $y = f(x) = mx + c = 5 x + 1$
1. Synthetic Data containing measurement error : `y`
    - $y_n = f(x_n) + \epsilon$ where $\epsilon \sim N(0,4^2)$
1. Minimize
    - $\min_{m,c}\sum_{n=1}^{30} ||y_n - f(x_n)||^2$

```python
import numpy as np
import matplotlib.pyplot as plt
f = np.poly1d([5, 1])
x = np.linspace(0, 10, 30)
y = f(x) + 4 * np.random.normal(size=len(x)) # add noise
xn = np.linspace(0, 10, 200)

m, c = np.polyfit(x, y, 1) # Linear regression
yn = np.polyval([m, c], xn)
plt.plot(x, y, 'or')
plt.plot(xn, yn)
plt.grid()
plt.legend(['data', 'regression model'])
plt.show()
```
